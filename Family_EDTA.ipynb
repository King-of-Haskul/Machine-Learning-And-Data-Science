```python
df1 = df[df.isna().any(axis=1)] #all rows containing null vales
df2 = df[df.isna().any(axis=1) == False] #all rows contianing non null values
```

```python
num_attribs = ["Total Household Income"]
cat_attribs = ['Main Source of Income', 'Agricultural Household indicator', 'Household Head Sex', 'Household Head Age', 'Household Head Marital Status',
       'Household Head Highest Grade Completed', 'Household Head Job or Business Indicator', 'Type of Household']
cat_attribs2 = ['Main Source of Income', 'Agricultural Household indicator', 'Household Head Sex', 'Household Head Age',
       'Household Head Highest Grade Completed', 'Household Head Job or Business Indicator']
y = df2[["Household Head Occupation", "Household Head Class of Worker"]]
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class nothing(BaseEstimator, TransformerMixin):
    def __init__(self):
         self.add = True
    
    def fit(self,X,y=None):
        return self
    
    def tranform(self, X, y=None):
        return X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('bruh', StandardScaler())
])
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs)
])
X = df2.copy()
X = full_pipeline.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn_clf = KNeighborsClassifier()
rnd_clf = RandomForestClassifier(n_estimators = 1000, max_leaf_nodes = 16, n_jobs = -1)

for clf in (knn_clf, rnd_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    n = accuracy_score(np.array(y_test)[:,0].reshape(-1,1), y_pred[:,0].reshape(-1,1))
    print(clf.__class__.__name__, n)
```

## Correlation Matrix and Plot

```python
corr = df.corr()
def plot(corr):
    mask = np.triu(np.ones_like(corr, dtype=bool))
    f, ax = plt.subplots(figsize=(15, 15))
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1.0, vmin=-1.0, center=0.0, square=True, linewidths=1, cbar_kws={"shrink": .5})
plot(corr)
```
